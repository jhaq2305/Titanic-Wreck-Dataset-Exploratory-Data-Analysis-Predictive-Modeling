```python
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv("Titanic-Dataset.csv")
```

```python
data.head()
```

```python
sns.heatmap(data.isnull())
```

```python
data.isnull().sum()
```

```python
data.drop(["Age"], axis=1, inplace=True)
data.drop(["Cabin"], axis=1, inplace=True)
data.drop(["Embarked"], axis=1, inplace=True)
```

```python
data.head()
```

```python
data.Sex = [1 if Sex == "male" else 0 for Sex in data.Sex]
data.head()
```

```python
data["Sex"].value_counts().plot(kind="bar")
```

```python
data.info()
```

```python
data['Name'] = data['Name'].astype('category').cat.codes
```

```python
data.head()
```

```python
data['Ticket'] = data['Ticket'].astype('category').cat.codes
```

```python
data.head()
```

```python
data.info()
```

```python
sns.heatmap(data.isnull())
```

```python
data.drop(["PassengerId"], axis=1, inplace=True)
```

```python
data.head()
```

```python
y = data["Survived"]
x = data.drop(["Survived"], axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
x_scaled
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.30, random_state=22)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train, y_train)
y_pred = lr.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(accuracy)
```

```python
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline([
    ("scaler", StandardScaler()),
    ("lr", LogisticRegression(max_iter=1000, solver="liblinear"))
])
```

```python
param_grid = {
    "lr__C": [0.01, 0.1, 1, 10, 100],
    "lr__penalty": ["l1", "l2"]
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(
    pipe,
    param_grid,
    cv=5,
    scoring="accuracy",
    n_jobs=-1
)
```

```python
grid.fit(x_train, y_train)

print("Best params:", grid.best_params_)
```

```python
from sklearn.metrics import classification_report
y_pred = grid.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
print("Test accuracy:", accuracy)
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix
conf_mat = confusion_matrix (y_test, y_pred)
conf_mat
```

```python
from sklearn.metrics import roc_auc_score, roc_curve
import matplotlib.pyplot as plt

y_score = lr.predict_proba(x_test)[:, 1]

logit_roc_auc = roc_auc_score(y_test, y_score)
fpr, tpr, thresholds = roc_curve(y_test, y_score)

plt.figure()
plt.plot(fpr, tpr, label='Logistic Regression (area = %0.2f)' % logit_roc_auc)
plt.plot([0, 1], [0, 1], 'r--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.savefig('Log_ROC')
plt.show()
```
