```python
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv("Titanic-Dataset.csv")
```

```python
data.head()
```

```python
sns.heatmap(data.isnull())
```

```python
data.isnull().sum()
```

```python
data.drop(["Age"], axis=1, inplace=True)
data.drop(["Cabin"], axis=1, inplace=True)
data.drop(["Embarked"], axis=1, inplace=True)
```

```python
data.head()
```

```python
data.Sex = [1 if Sex == "male" else 0 for Sex in data.Sex]
data.head()
```

```python
data["Sex"].value_counts().plot(kind="bar")
```

```python
data.info()
```

```python
data['Name'] = data['Name'].astype('category').cat.codes
```

```python
data.head()
```

```python
data['Ticket'] = data['Ticket'].astype('category').cat.codes
```

```python
data.head()
```

```python
data.info()
```

```python
sns.heatmap(data.isnull())
```

```python
data.drop(["PassengerId"], axis=1, inplace=True)
```

```python
data.head()
```

```python
y = data["Sex"]
x = data.drop(["Sex"], axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
x_scaled
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.30, random_state=22)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train, y_train)
y_pred = lr.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(accuracy)
```
